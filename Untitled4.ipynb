```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Carregar dados
df = pd.read_csv('/content/ecommerce_estatistica - ecommerce_estatistica.csv.csv')

# 2. Limpeza (sempre!)
print(f"Linhas originais: {len(df)}")

# Remover valores vazios
df = df.dropna(subset=['Nota', 'Preço', 'Desconto'])

# Histogrma
plt.figure(figsize=(10, 6))
sns.histplot(df['Preço'], bins=20, color='blue', alpha=0.8) # Changed sns.hist to sns.histplot for clarity and modern usage
plt.title('Histograma de Preços')
plt.xlabel('Preço')
plt.ylabel('Frequência')
plt.show()

# Barras - Melhores Marcas
melhores_marcas = df.groupby('Marca')['Nota'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(12, 6))
melhores_marcas.plot(kind='bar', color='green', edgecolor='black')
plt.title('Top 10 Marcas com melhores avaliações', fontsize=14, fontweight='bold')
plt.xlabel('Marca')
plt.ylabel('Média de Nota')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Regressão: Preço X Nota
plt.figure(figsize=(12,6))
sns.regplot(x='Preço', y='Nota', data=df, scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.title('Relação entre Preço e Nota', fontsize=14, fontweight='bold')
plt.xlabel('Preço (R$)')
plt.ylabel('Nota')
plt.show()

# Boxplot: Distibuição de preço por gênero
# Preparar dados com nomes curtos
df['Gênero_Curto'] = df['Gênero'].map({
    'Masculino': 'Masc',
    'Feminino': 'Fem',
    'Sem gênero': 'Neutro',
    'Sem gênero infantil': 'Infantil',
    'Meninos': 'Meninos',
    'Meninas': 'Meninas',
    'Bebês': 'Bebês',
    'Unissex': 'Unissex'
})

# Remover gêneros muito raros (opcional)
contagem = df['Gênero_Curto'].value_counts()
generos_principais = contagem[contagem > 5].index
df_filtrado = df[df['Gênero_Curto'].isin(generos_principais)]

# Gráfico
plt.figure(figsize=(12,6))
sns.boxplot(x='Gênero_Curto', y='Preço', data=df_filtrado, palette='Set2')
plt.title('Distribuição de Preços por Gênero', fontsize=14, fontweight='bold')
plt.xlabel('Gênero', fontsize=12)
plt.ylabel('Preço (R$)', fontsize=12)
plt.tight_layout()
plt.show()

# Análise estatística
print("Estatísticas por Gênero:")
print(df_filtrado.groupby('Gênero_Curto')['Preço'].describe().round(2))

# Pizza: Material mais usado
material_counts = df['Material'].value_counts().head(6)
plt.figure(figsize=(10,8))
plt.pie(material_counts, labels=material_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribuição por Tipo de Material', fontsize=14, fontweight='bold')
plt.show()
```
